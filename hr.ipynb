## П.5. 
С использованием t-test статистики проверить взаимосвязи между несколькими переменными.

```python

df.hist(edgecolor='black', linewidth=1.2)
fig=plt.gcf()
fig.set_size_inches(12,6)
plt.show()
```

```python
df['sales'].value_counts()
```

```python
# Рассмотрим две выборки: среднее время проводимое на работе сотрудниками отдела IT и отдела продаж
IT_average_montly_hours = df[df['sales'] == 'IT'][['average_montly_hours']]
```

```python
sales_average_montly_hours = df[df['sales'] == 'sales'][['average_montly_hours']]
```

```python
IT_average_montly_hours.shape
```

```python
sales_average_montly_hours.shape
```

```python
sts.ttest_ind(IT_average_montly_hours, sales_average_montly_hours)
```

p-value выше значимого, сдедовательно мы не можем отвергнуть нулевую гипотезу и средние значеня двух выборок равны

```python
# Рассмотрим две выборки: уровень удовлетворенноти работой у сотрудников отдела IT и отдела продаж
IT_satisfaction_level = df[df['sales'] == 'IT'][['satisfaction_level']]
```

```python
sales_satisfaction_level = df[df['sales'] == 'sales'][['satisfaction_level']]
```

```python
sts.ttest_ind(IT_satisfaction_level, sales_satisfaction_level)
```

p-value выше значимого, сдедовательно мы не можем отвергнуть нулевую гипотезу и средние значеня двух выборок равны

```python
# Рассмотрим две выборки: уровень удовлетворенноти работой у сотрудников с низкой зарплаттой и высокой
sat_low_salary = df[df['salary'] == 'low'][['satisfaction_level']]
```

```python
sat_high_salary = df[df['salary'] == 'high'][['satisfaction_level']]
```

```python
sts.ttest_ind(sat_low_salary, sat_high_salary)
```

p-value ниже значимого, сдедовательно мы отвергаем нулевую гипотезу и выборки имеют разные средние

## П.10
Проверить несколько гипотез для данных,чтобы понять почему сотрудники покидают компанию.

```python
left = df[df['left'] == 1][['satisfaction_level']]
```

```python
stay = df[df['left'] == 0][['satisfaction_level']]
```

```python
left.mean()
```

```python
stay.mean()
```

```python
sts.ttest_ind (stay, left)
```

p-value ниже значимого, сдедовательно мы отвергаем нулевую гипотезу и выборки имеют разные средние
средний уровень удовлетворенности работой у уволевшихся ниже, следовательно можно предположить что это влияет на решение об уходе
